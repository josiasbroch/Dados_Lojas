**Passo a passo:**

```python
#Lógica de programação:

#Passo 0 - Entender o desafio que você quer resolver

#Passo 1 - Percorrer todos os arquivos de pasta base de dados (Pasta Vendas)
import os
import pandas as pd

lista_arquivos = os.listdir("/content/drive/MyDrive/Curso basico de python/Vendas/Vendas") #lista todos os arquivos q estao no diretorio
display(lista_arquivos)
```

```python
tabela_total = pd.DataFrame()

#Passo 2 - Importar as bases de dados de vendas
for arquivo in lista_arquivos:
  #se tem "Vendas no nome do arquivo, então"
  if "Vendas" in arquivo:
    #importar o arquivo
    tabela = pd.read_csv(f"/content/drive/MyDrive/Curso basico de python/Vendas/Vendas/{arquivo}")
    tabela_total = tabela_total.append(tabela) #append = adicionar
    
#Passo 3 - Tratar/Compilar as bases de dados
display(tabela_total)
```

```python
# Passo 4 - Calcular o produto mais vendido (em quantidade)
tabela_produtos = tabela_total.groupby('Produto').sum()
tabela_produtos = tabela_produtos[["Quantidade Vendida"]].sort_values(by="Quantidade Vendida", ascending=False) 
#sort_values() = organiza os valores, ascending=True : Forma crescente, ascending=False : Forma decrescente
display(tabela_produtos) 
```

```python
# Passo 5 - Calcular o produto que mais faturou (em faturamento)
tabela_total['Faturamento'] = tabela_total['Quantidade Vendida'] * tabela_total['Preco Unitario']

tabela_faturamento = tabela_total.groupby("Produto")
tabela_faturamento = tabela_faturamento[["Faturamento"]].sum().sort_values(by="Faturamento", ascending=False) 
display(tabela_faturamento) 
```

```python
# Passo 6 - Calcular a loja/cidade que mais vendeu (em faturamento) 
tabela_lojas = tabela_total.groupby('Loja').sum()
tabela_lojas = tabela_lojas[["Faturamento"]].sort_values(by="Faturamento", ascending=False)
display(tabela_lojas) 
```

```python
# Passo 7 - Criar um gráfico/dashboard da loja/cidade que mais vendeu
import plotly.express as px
grafico = px.bar(tabela_lojas, x=tabela_lojas.index, y="Faturamento")
#É tabela_lojas.index e não "Loja" pq a loja é um indice eo faturamento nao
grafico.show() 

```
